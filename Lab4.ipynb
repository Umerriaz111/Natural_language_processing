```python
import nltk
```

```python
paragraph="This, is simple text for generating bigrams. this "
```

```python
from nltk.tokenize import word_tokenize
```

```python
sentence=word_tokenize(paragraph)
```

```python
len(sentence)
```

```python
sentence
```

# Bigrams

```python
list1=[]
for w in range(len(sentence)-1):
    list1.append(sentence[w]+" "+sentence[w+1])
    
```

```python
list1
```

# Trigrams

```python
list2=[]
for w in range(len(sentence)-2):
    list2.append(sentence[w]+" "+sentence[w+1]+" "+sentence[w+2])
```

```python
list2
```

# Pos tags for bigrams

```python

    pos_tags=nltk.pos_tag(list1)
   
```

```python
pos_tags
```

```python
all_words=nltk.FreqDist()
for w in sentence:
    all_words[w.lower()]=all_words[w.lower()]+1
```

```python
all_words
```

```python
pos_tags[0]
```

```python
from collections import Counter
```

```python
freq = Counter(list1)
```

```python
for ngram, count in freq.items():
    print(ngram, count)
```

```python
import matplotlib.pyplot as plt
```

```python
x = [str(ngram) for ngram in freq.keys()]
y = list(freq.values())
plt.bar(x, y)
plt.show()
```
