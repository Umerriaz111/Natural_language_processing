# Part 2

# 2.1

# Tokenization

```python
words_tokens_pos=[]
for tweets in postive_tweets:
    X1=word_tokenize(tweets)
    words_tokens_pos.append(X1)
```

```python
words_tokens_Neg=[]
for tweets in Negative_tweets:
    X1=word_tokenize(tweets)
    words_tokens_Neg.append(X1)
```

```python
words_tokens_Neu=[]
for tweets in Neutral_tweets:
    X1=word_tokenize(tweets)
    words_tokens_Neu.append(X1)
```

# 2.2

# StopWords removal

```python
stop_words_filtering_pos=[]
for lists in words_tokens_pos:
    for Words in lists:
        if Words not in set(stopwords.words()):
            stop_words_filtering_pos.append(Words)
       
    
```

```python
stop_words_filtering_Neg=[]
for lists in words_tokens_Neg:
    for Words in lists:
        if Words not in set(stopwords.words()):
            stop_words_filtering_Neg.append(Words)
```

```python
stop_words_filtering_Neu=[]
for lists in words_tokens_Neu:
    for Words in lists:
        if Words not in set(stopwords.words()):
            stop_words_filtering_Neu.append(Words)
```

```python
alltext=set(postive_tweets+Negative_tweets+Neutral_tweets)
```

```python
alltext
```

# 2.3

# Normalization

```python
all_text=[]
for words in alltext:
    X=words.lower()
    all_text.append(X)
  
```

```python
all_text
```

# 2.4

# Extraction of Words starting with # and @

```python
Tokens_list=[]
for Sentence in all_text:
    Tokens_list.append(Sentence.split(" "))
```

```python
Stop_words_Removal=[]
for lists in Tokens_list:
    for words in lists:
         if words not in set(stopwords.words("english")):
                Stop_words_Removal.append(words)
    
    
```

```python
Hastages=[]
usersnames=[]
for words in Stop_words_Removal:
    if words.startswith("#"):
        Hastages.append(words)
    elif words.startswith("@"):
        usersnames.append(words)
        
   
```

```python
dictionary={}
dictionary['Hastages']=Hastages
dictionary['usersnames']=usersnames
dictionary['Text']=Stop_words_Removal
```

```python
URL=[]
for words in Stop_words_Removal:
    if words.startswith("https"):
        URL.append(words)
    
```

```python
dictionary['URL']=URL
```

# 2.6

# Removing Emojis

```python
Clean_text_from_emojis=[]
# Define a regular expression pattern to match emoticons/emojis
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002702-\U000027B0"  # other miscellaneous symbols
        u"\U000024C2-\U0001F251" 
                           "]+", flags=re.UNICODE)
for w in Stop_words_Removal:
     Clean_text_from_emojis.append(emoji_pattern.sub(r'', w))
```
